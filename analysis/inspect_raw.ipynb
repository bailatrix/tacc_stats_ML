```python
job_dfs = {}

for job,cpi in notable.items():
    filename = job[1]+'_'+job[0]+'.csv'
    try:
        job_dfs[job] = pd.read_csv( source_dir+filename, index_col=[0,1,2], low_memory=False )
    except:
        print "Jobid", job[0], "has not been processed"
```
