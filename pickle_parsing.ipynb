```python
# dependencies
from tacc_stats.pickler.job_stats import Job
import cPickle as pickle
import pandas as pd
import numpy as np
import argparse
from os import listdir
```

```python
# directory of all jobs on given date
source_dir = '/oasis/projects/nsf/sys200/tcooper/xsede_stats/comet_pickles/2016-08-21/'

# get list of files in directory
pickled_jobs = [ source_dir+job for job in listdir( source_dir ) ]
jobs = []

# access and open 100 valid jobs
# append to list jobs
count = 0
for job in pickled_jobs:
    if count < 100:
        try:
            pickle_file = open(job, 'rb')
            jobid = pickle.load(pickle_file)
            jobs.append(jobid)
            pickle_file.close()
            count += 1
        except:
            next
    else:
        break
```

```python
stats_dict = {}

# get stats from hosts
for job in jobs:
    for host_name, host in jobs[0].hosts.iteritems():
        host.get_stats_paths
        host.read_stats_file_header
        host.parse_stats
        host.read_stats_file
        host.gather_stats
        host.get_stats
        stats_dict.update(host.stats)
```

```python
# merge all data
df = pd.DataFrame( stats_dict )
df = df.reset_index()
```

```python
df = df[["index","intel_hsw","ib",
         # "ib_sw","ib_ext",
        "llite","lnet",
         #"mdc",
         "osc","block",
        "cpu","mem","net",
         "nfs","numa","proc",
        "sysv_shm","tmpfs"]]
df.head()
```

```python
# compute avg of collected values
for row in range(0, len(df.index)):
    for col in range(0, len(df.columns)):
        cell = df.iat[row,col]
        if ( type(cell) == np.ndarray ):
            df.iat[row,col] = np.mean(cell)

df.head()
```
