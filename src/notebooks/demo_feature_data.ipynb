### Preview
- Counts of features represented in sample
- Note: This may or may not represent the full set of combinations of stat, device, metric seen in the population

```python
features = [ item for item in listdir('./data/') if item != '.ipynb_checkpoints' ]
stats = {}
devices = {}
metrics = {}

for item in features:
    stat = item.split('--')[0]
    cut_idx = item.rfind('--')
    dev = item[: cut_idx]
    metric = item.split('--')[2]
    
    if stat not in stats:
        stats[ stat ] = 1
    else:
        stats[ stat ] += 1
        
    if dev not in devices:
        devices[ dev ] = 1
    else:
        devices[ dev ] += 1
        
    if metric not in metrics:
        metrics[ metric ] = 1
    else:
        metrics[ metric ] += 1

len(features)
```
